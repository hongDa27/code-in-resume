```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib
%matplotlib inline
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_excel('05-19data/制造业工业表.xlsx')
```

```python
df
```

```python
df2 = pd.read_excel('05-19data/washed_data.xlsx',index_col=0)
```

```python
data = pd.merge(left = df,right = df2,on = 'year',how = 'outer')
```

```python
data.drop(columns = '学历年龄',inplace=True)
```

```python
len(data.columns)
```

```python
data['ckb'] = data['資本サービス投入指数']*data['実質資本ストック指数']
```

```python
data['total_working_time'] = data['所定内実労働時間数'] + data['超過実労働時間数']
```

```python
data['salary'] = data['きまって支給する現金給与額(千円)'] *12 +data['年間賞与その他特別給与額(千円)']
```

```python
data['勤続年数2乘'] = np.square(data['勤続年数'])
```

```python
data = data.astype(np.float64)
```

```python
data['実際中间投入'] = np.log(data['中间投入'])
```

```python
data['実質産出'] = np.log(data['出荷额'])
```

```python
 male_data = data[data['sex']==1]
```

```python
male_data.reset_index(drop = 'index',inplace=True)
```

```python
female_data = data[data['sex'] ==0]
```

```python
female_data.reset_index(inplace=True,drop = 'index')
```

```python
female_data
```

# 女性モデル

```python
female_data.head()
```

```python
y = female_data['実質産出']
```

```python
feature  = female_data.loc[:,['勤続年数','勤続年数2乘','education','実際中间投入','ckb','total_working_time']]
```

```python
feature = sm.add_constant(feature)
```

```python
res = sm.OLS(y,feature).fit()
```

```python
res.summary()
```
